# Multi-output

```python
from models.cpmp_transformer_v6 import CPMPTransformer
from training.training import load_model

model_name = "multi"
model = load_model(CPMPTransformer, model_name)
```

```python
from generation.data import get_feasible_moves
from generation.instances import read_instance
from solvers.model import ModelSolver
from generation.adapters import EnrichedStackMatrixAdapter, StackMatrix4D3FAdapter, ExtraDataAdapter3F
import copy

def get_child_layouts(layout):
    moves = get_feasible_moves(layout)

    lay_copies = []
    for (i, j) in moves:
        lay_copy = copy.deepcopy(layout)
        lay_copy.move(i, j)
        lay_copies.append(lay_copy)

    return lay_copies

model_solver = ModelSolver(model, EnrichedStackMatrixAdapter(StackMatrix4D3FAdapter, ExtraDataAdapter3F))

instance_file = "E5-25-5/E5-25-5-8.txt"
H = 7
max_steps = 500

layout = read_instance(instance_file, H)
child_layouts = get_child_layouts(layout)
print(len(child_layouts))

results = model_solver.solve_from_layouts(child_layouts, H, max_steps)
print(results)
```
